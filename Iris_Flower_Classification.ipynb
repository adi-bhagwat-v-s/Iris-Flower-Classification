```python
data = df.values
x = data[:, 0:4]
y = data[:, 4]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
len(y_test)
```

Support Vector Machine Algorithm

```python
from sklearn.svm import SVC
model_svc = SVC()
model_svc.fit(x_train, y_train)
```

```python
prediction = model_svc.predict(x_test)
prediction
```

```python
from sklearn.metrics import accuracy_score
score_svm = accuracy_score(y_test, prediction)
print("svm accuracy => ", score_svm*100)
```

```python
for i in range((len(prediction))):
  print(y_test[i], prediction[i])
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model_lr = LogisticRegression()
model_lr.fit(x_train, y_train)
```

```python
prediction_lr = model_lr.predict(x_test)
score_lr = accuracy_score(y_test, prediction_lr)
print("logistic regression => ", score_lr*100)
```

```python
for i in range((len(prediction))):
  print(y_test[i], prediction[i])
```

Decison Tree

```python
from sklearn.tree import DecisionTreeClassifier
model_dt = DecisionTreeClassifier()
model_dt.fit(x_train, y_train)
```

```python
prediction_dt = model_dt.predict(x_test)
score_dt = accuracy_score(prediction_dt, y_test)
print("decision tree score => ", score_dt*100)
```
